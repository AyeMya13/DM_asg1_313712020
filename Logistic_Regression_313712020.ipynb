```python
%matplotlib inline
%reload_ext autoreload
%autoreload 2

import numpy as np
import matplotlib.pyplot as plt

from model.metrics import logloss, evaluate_binary_classifier
from model.gradients import logloss_sigmoid_grad
from model.utils import *
from model.activations import sigmoid
from model.linear_model import LinearModel
```

```python
# Helper functions
seed = 42  # do not change

def get_loss_history(model):
    # Return the loss history from the model.
    for attr in ["losses", "train_losses", "loss_history", "history", "loss"]:
        if hasattr(model, attr):
            value = getattr(model, attr)
            if isinstance(value, (list, tuple, np.ndarray)):
                return np.asarray(value).ravel()
    raise AttributeError(
        "Cannot find loss history in model. "
        "Please check whether your LinearModel stores it as "
        "'losses', 'train_losses', or another attribute."
    )

def train_logistic_model(X, y, lr=0.01, n_iteration=500):
    np.random.seed(seed)

    loss_fn = logloss
    act_fn = sigmoid
    grad_fn = logloss_sigmoid_grad

    X_train, y_train, X_val, y_val = get_train_val(X, y)

    dim = X.shape[1] - 1
    model = LinearModel(dim, False, loss_fn, grad_fn, act_fn)
    model.fit(X_train, y_train, lr=lr, n_iteration=n_iteration, val_ratio=0.2)

    return model, X_train, y_train, X_val, y_val

def plot_decision_boundary(model, X_val, y_val, title):
    plt.figure(figsize=(5, 4))
    plt.scatter(X_val[:, 1], X_val[:, 2], c=y_val)

    ex1 = np.linspace(np.min(X_val[:, 1]) - 0.5, np.max(X_val[:, 1]) + 0.5, 100)
    ex2 = -(model.W[1] * ex1 + model.W[0]) / model.W[2]

    plt.plot(ex1, ex2, color='r', label='decision boundary')
    plt.title(title)
    plt.legend()
    plt.show()
```

```python
# 2(a), 2(b): Learning-rate experiments for all datasets

datasets = ["A", "B", "C", "D"]
learning_rates = [0.1, 0.01, 0.001]

for ds in datasets:
    print("=" * 80)
    print(f"Dataset: logistic_data_{ds}")
    print("=" * 80)

    loaded_data = np.load(f"data/logistic_data/logistic_data_{ds}.npz")
    X = loaded_data["X_data"]
    y = loaded_data["y_data"]
    W = loaded_data["W_data"]

    print(f"X size: {X.shape}, y size: {y.shape}")

    for lr in learning_rates:
        print("-" * 80)
        print(f"Learning rate = {lr}")

        model, X_train, y_train, X_val, y_val = train_logistic_model(
            X, y, lr=lr, n_iteration=500
        )

        plt.figure(figsize=(6, 4))
        plt.plot(get_loss_history(model))
        plt.title(f"logistic_data_{ds} - Loss Curve (lr={lr})")
        plt.xlabel("Iteration")
        plt.ylabel("Loss")
        plt.grid(True)
        plt.show()

        y_val_pred = model.predict(X_val)
        evaluate_binary_classifier(
            y_true=y_val,
            y_pred=y_val_pred,
            title=f"logistic_data_{ds} Evaluation (lr={lr})"
        )

        plot_decision_boundary(
            model, X_val, y_val,
            title=f"logistic_data_{ds} - Decision Boundary (lr={lr})"
        )
```

```python
# 2(d), 2(e): Iteration experiments for all datasets (fixed lr = 0.01)

datasets = ["A", "B", "C", "D"]
iteration_list = [500, 1000, 1500]
fixed_lr = 0.01

for ds in datasets:
    print("=" * 80)
    print(f"Dataset: logistic_data_{ds}")
    print("=" * 80)

    loaded_data = np.load(f"data/logistic_data/logistic_data_{ds}.npz")
    X = loaded_data["X_data"]
    y = loaded_data["y_data"]
    W = loaded_data["W_data"]

    print(f"X size: {X.shape}, y size: {y.shape}")

    for n_iter in iteration_list:
        print("-" * 80)
        print(f"n_iteration = {n_iter}")

        model, X_train, y_train, X_val, y_val = train_logistic_model(
            X, y, lr=fixed_lr, n_iteration=n_iter
        )

        plt.figure(figsize=(6, 4))
        plt.plot(get_loss_history(model))
        plt.title(f"logistic_data_{ds} - Loss Curve (n_iteration={n_iter})")
        plt.xlabel("Iteration")
        plt.ylabel("Loss")
        plt.grid(True)
        plt.show()

        y_val_pred = model.predict(X_val)
        evaluate_binary_classifier(
            y_true=y_val,
            y_pred=y_val_pred,
            title=f"logistic_data_{ds} Evaluation (n_iteration={n_iter})"
        )

        plot_decision_boundary(
            model, X_val, y_val,
            title=f"logistic_data_{ds} - Decision Boundary (n_iteration={n_iter})"
        )
```
